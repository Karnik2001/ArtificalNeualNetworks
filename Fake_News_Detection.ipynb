```python
# import all of the necessary libraries
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
```

```python
# load the dataset
# display first 20 rows 
fake = pd.read_csv('/content/fake.csv')
fake.head(20)
```

```python
# select necessary columns for model
fake = fake[["text","type"]]

# create feature and target variables
x = np.array(fake["text"])
y = np.array(fake["type"])
```

```python
# convert text data into numerical form
fv = CountVectorizer()
X = fv.fit_transform(fake["text"].values.astype('U'))
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.33, random_state =42)

flf = MultinomialNB()
flf.fit(X_train,y_train)
```

```python
# predict the type of text
fake_sample = input('Enter a text')
fake = fv.transform([fake_sample]).toarray()
print(flf.predict(fake))
```
